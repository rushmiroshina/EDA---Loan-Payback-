## Data Cleaning

### Taking care of NULL Columns

```python
#checking the percentage of null values in the null columns
app_null=(app.isnull().sum()*100)/len(app)
app_null
```

```python
#Filtering the Null columns having more than 40% NULL values
app_null=app_null[app_null>=40]
app_null
```

These columns seems to be useless,hence dropping the columns

```python
#dropping columns with more than 40% null values
app_1=app.drop(app_null.index,axis=1)
```

```python
app_1.shape
```

The shape of the column has been changed after the removal of columns with more than 40% NULL values

```python
app_1.info()
```

OCCUPATION_TYPE seems to have lot of NULL values and this column seems to be useful,so this has to be fixed 

```python
#comparing OCCUPATION_TYPE and AMT_INCOME_TOTAL
plt.figure(figsize=[15,8])
plt.tick_params(rotation=90)
sns.boxplot(data=app_1[app_1.AMT_INCOME_TOTAL < 10000000], x='OCCUPATION_TYPE', y='AMT_INCOME_TOTAL')
plt.show()
```

There seems to be lot of outliers,so substituting the NULL values with mean is not advisable.It can be substituted with Mode of the column

```python
#Substituting NULL values of OCCUPATION_TYPE with mode of the column
mode_occ = app_1.OCCUPATION_TYPE.mode()[0]
app_1.OCCUPATION_TYPE.fillna(value=mode_occ, inplace=True)
```

AMT_ANNUITY and AMT_GOODS_PRICE may be valuable and the null values has to be taken care

These columns refers some amount,hence substitution of mean would be inappropriate 

```python
#Taking care of NULL values for AMT_GOODS_PRICE and AMT_ANNUITY
mode_7 = app_1.AMT_GOODS_PRICE.mode()[0]
app_1.AMT_GOODS_PRICE.fillna(value=mode_7, inplace=True)

mode_8 = app_1.AMT_ANNUITY.mode()[0]
app_1.AMT_ANNUITY.fillna(value=mode_8, inplace=True)
```

CNT_FAM_MEMBERS seems to have only 2 NULL values,hence the null rows can be removed

```python
#dropping NULL rows for the column 
app_1 = app_1.dropna(subset=['CNT_FAM_MEMBERS'])
```

```python
#checking if the rows has been dropped
app_1.info()
```

'AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_DAY', 'AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_QRT' also has many null values.This also can be substituted by mode

```python
#Mode substitute for 'AMT_REQ_CREDIT_BUREAU_MON', 'AMT_REQ_CREDIT_BUREAU_WEEK', 'AMT_REQ_CREDIT_BUREAU_DAY', 'AMT_REQ_CREDIT_BUREAU_HOUR', 'AMT_REQ_CREDIT_BUREAU_QRT'
mode_1 = app_1.AMT_REQ_CREDIT_BUREAU_MON.mode()[0]
app_1.AMT_REQ_CREDIT_BUREAU_MON.fillna(value=mode_1, inplace=True)

mode_2 = app_1.AMT_REQ_CREDIT_BUREAU_WEEK.mode()[0]
app_1.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(value=mode_2, inplace=True)

mode_3 = app_1.AMT_REQ_CREDIT_BUREAU_DAY.mode()[0]
app_1.AMT_REQ_CREDIT_BUREAU_DAY.fillna(value=mode_3, inplace=True)

mode_4 = app_1.AMT_REQ_CREDIT_BUREAU_HOUR.mode()[0]
app_1.AMT_REQ_CREDIT_BUREAU_HOUR.fillna(value=mode_4, inplace=True)

mode_5 = app_1.AMT_REQ_CREDIT_BUREAU_QRT.mode()[0]
app_1.AMT_REQ_CREDIT_BUREAU_QRT.fillna(value=mode_5, inplace=True)

mode_6 = app_1.AMT_REQ_CREDIT_BUREAU_YEAR.mode()[0]
app_1.AMT_REQ_CREDIT_BUREAU_YEAR.fillna(value=mode_6, inplace=True)
```

```python
#checking the updated dataset before doing further cleaning
app_1.head(10)
```

### Handling inconsistent columns


The Days columns seems to have some inappropriate data

```python
#Handling Negative values 
app_1.DAYS_BIRTH=abs(app_1.DAYS_BIRTH)
app_1.DAYS_EMPLOYED=abs(app_1.DAYS_EMPLOYED)
app_1.DAYS_REGISTRATION=abs(app_1.DAYS_REGISTRATION)
app_1.DAYS_ID_PUBLISH=abs(app_1.DAYS_ID_PUBLISH)
```

```python
#converting days to years
app_1.DAYS_BIRTH=app_1.DAYS_BIRTH/365
app_1.DAYS_EMPLOYED=app_1.DAYS_EMPLOYED/365
app_1.DAYS_REGISTRATION=app_1.DAYS_REGISTRATION/365
app_1.DAYS_ID_PUBLISH=app_1.DAYS_ID_PUBLISH/365
```

### Data Cleaning

```python
#checking NULL values
null_per=prev.isnull().sum()*100/len(prev)
null_per
```

```python
null_per=null_per[null_per>50]
null_per
```

```python
prev_1=prev.drop(null_per.index,axis=1)
```

```python
prev_1.shape
```

```python
#correcting date
def date_cor(prev):
    days = [d for d in prev if d.startswith('DAYS')]
    for d in days:
        prev[d] = abs(prev[d])
    return prev
    
```

```python
prev_1=date_cor(prev_1)
```

```python

prev_1
```

```python
#taking care of inconsistent data
prev_1=prev_1.replace("XNA",np.NaN)
prev_1=prev_1.replace("XAP",np.NaN)
```
